```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# OR Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# OR Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

### BUBBLE CHART

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*4) #the no 4 is use to set the ballsize. you can choos any number
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# using list comprehension to change the ball size
#The format for list comprehension is as follows.

#new_list = [expression for item in list if conditional]

plt.scatter(x_axis, y_axis, s = [i * 4 for i in y_axis])
```

```python
#with object oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
ax.scatter(x_axis, y_axis, s = [i * 4 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
ax.scatter(x_axis, y_axis, marker="0", facecolors="blue", edgecolors="black", label='Boston', s = [i * 5 for i in y_axis])
#plt.scatter(x_axis, y_axis, marker="0", facecolors="blue", edgecolors="black", label='Boston', s = [i * 5 for i in y_axis])
# Create labels for the x and y axes.
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add the legend.
ax.set_legend()
#Add a limit for the x-axis data.
ax.set_xlim(11,26)
#plt.gca().invert_yaxis()
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0.5)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
